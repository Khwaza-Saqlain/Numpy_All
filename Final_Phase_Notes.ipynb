## Advances operations with Business examples

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data structure: [shop_id, 2021, 2022, 2023, 2024]
sales_data = np.array([
    [1, 25000, 30000, 33000, 36000],   # Book World
    [2, 28000, 32000, 34000, 37000],   # Toy Planet
    [3, 22000, 25500, 27000, 31500],   # Gadget Garage
    [4, 19000, 23000, 24500, 26500],   # Fashion Central
    [5, 33000, 35000, 37000, 40000]    # Sports Hub
])

print("===========Store Sales Analysis==========")
print(sales_data)
```

```python
#First step should be to identify the shape of the data that you are working with
print(np.shape(sales_data)) # we got the result as (5, 5) which means its a 5x5 matrix
```

```python
#Get the data of first 3 stores
first_3_stores = sales_data[:3]
# print("First 3 stores : ",first_3_stores)
print(sales_data[:,1:]) # so here 1st parameter represents the row and second parameter represents the column and will be manipulated accordingly.
```

```python
#calculate yearly total of the stores
yearly_total = np.sum(sales_data[:,1:], axis=1) #so axis 0 means column wise calculation, and axis 1 means row wise calculation.
print(yearly_total)
```

```python
print("Yearly_total sum = ", np.sum(yearly_total))
```

```python
#minimum sales per store
min_sales_per_store = np.min(sales_data[:,1:],axis=1) #store wise calculation is done, meaning row wise calculation of the matrix.
print(min_sales_per_store)
```

```python
#maximum sales for any store
max_sales  = np.max(sales_data[:, 1:], axis=0)
max_amount = np.max(max_sales)
print("Max Sales per year : ",max_sales)
print("Max Sales Number :", max_amount)
```

```python
#Now find out the cumulative sum for each store over the years
cumsum = np.cumsum(sales_data[:,1:], axis=1)
print("The cumulative sum row wise : \n", cumsum)
```
