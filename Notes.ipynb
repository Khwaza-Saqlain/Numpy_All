### Comparing speed of List and Numpy Array

```python
import time 
start = time.time()
py_list = [i*2 for i in range(1000000)]
end = time.time()
print("List Process completion time : ", end-start)

start = time.time()
np_array = np.arange(1000000) * 2
end = time.time()
print("np_array Process completion time : ", end-start)

```

## Creating Array from scratch

```python
#Filling the matrix with zeros
zeros = np.zeros((3,4))
print(zeros)

#filling the matrix with ones
ones = np.ones((3,2))
print(ones)

#filling up the matrix with the desired number
num = np.full((2,3),7)
print(num)

#filling up with random values
random_matrix = np.random.random((2,3))
print(random_matrix)

#making up an array
sequence = np.arange(0,20,2) #the parameters are start, stop, step and it doesn't include the outer bound
print(sequence)
```

## Vector, Matrix and Tensor

```python
#Vector
vector = np.array([1,2,3]) #vector is like an array , 1 dimentional
print("vector : ", vector)

matrix = np.array([[1,2,3],
                   [3,4,5]]) #matrix is always a 2 dimentional array.
print("matrix : ", matrix)

tensor = np.array([[[1,2,3], [3,2,5]],
                    [[3,4,5],[5,3,9]]]) #tensor is always multidimentional , anything more than 2 dimentional is termed as tensor.
print("tensor : ", tensor)
```

## Array Properties

```python
array = np.array([[2,4,5],[5,5,2]])
print("Shape : ", array.shape)
print("Dimension : ", array.ndim)
print("Size : ", array.size)
print("Data Type of the array : ", array.dtype)
```

## Array Reshaping

```python
arr = np.arange(12)
print("Original Array : ",arr)

reshaped = arr.reshape((3,4))
print("Reshaped Array : \n", reshaped) #This will reshape the original array, in this case it is reshaped to a 2D array

flattened = reshaped.flatten()
print("Matrix flattened : \n", flattened) #This will flatten the array, always returns the copy of the array.

raveled = reshaped.ravel()
print("Raveled : \n", raveled) #This always returns the original array modified (In case of contigous array) if the array is non contigous then it will return a copy .

# Transpose means in a matrix the rows will become columns
transpose = reshaped.T
print("Transposed Matrix : \n", transpose)
```
